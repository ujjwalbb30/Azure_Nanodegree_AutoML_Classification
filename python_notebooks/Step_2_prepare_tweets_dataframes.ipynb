# **Preparing different Tweets dataframes for all the companies**

### Now that we have explored all the datasets, we will prepare following types of dataframe for all the companies:    

### (1) Same Day's aggregated tweets dataframe: It will contain aggregated value of tweets and its attributes for the current day.

### (2) Previous Day's aggregated tweets dataframe: It will contain aggregated value of tweets and its attributes for the previous day.

### (3) Last 3 Days' aggregated tweets dataframe: It will contain aggregated value of tweets and its attributes for the last 3 days.

### (4) Last 7 Days' aggregated tweets dataframe: It will contain aggregated value of tweets and its attributes for the last 7 days.

### We will filter the main dataframes using tweet_id associated with each company to create separate dataframes for each company. Then we will aggregate the data as per the above configurations and export it for further use.

# **For Apple (AAPL)**

```python
# importing required dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
# mounting the google drive to access datasets
from google.colab import drive 
drive.mount('/content/gdrive')
```

```python
# importing the 'Company.csv' dataset
company_name_ts = pd.read_csv('/content/gdrive/MyDrive/DATA_602_Project/datasets/Company.csv')
company_name_ts
```

```python
# importing the 'Company_Tweet.csv' dataset
company_tweet_id_ts = pd.read_csv('/content/gdrive/MyDrive/DATA_602_Project/datasets/Company_Tweet.csv')
company_tweet_id_ts
```

```python
# filtering 'company_tweet_id_ts' to get the tweet_id associated with apple ('AAPL')
apple_tweet_id = company_tweet_id_ts[company_tweet_id_ts['ticker_symbol']=='AAPL']
apple_tweet_id
```

```python
# converting the 'tweet_id' column of filtered dataset to a list
apple_tweet_ids = list(apple_tweet_id['tweet_id'])
```

```python
# importing the 'Tweet.csv' dataset that contains tweet details.
company_tweets = pd.read_csv('/content/gdrive/MyDrive/DATA_602_Project/datasets/Tweet.csv')
company_tweets
```

```python
# filtering 'company_tweets' using 'apple_tweet_ids'
apple_tweets_df = company_tweets[company_tweets.tweet_id.isin(apple_tweet_ids)]

# resetting the index and removing the 'tweet_id' column
apple_tweets_df = apple_tweets_df.reset_index(drop=True)
apple_tweets_df = apple_tweets_df.drop(columns=['tweet_id'])
apple_tweets_df
```

```python
# converting the 'post_date' column to required datetime format and eastern time
apple_tweets_df['date'] = pd.to_datetime(apple_tweets_df['post_date'],unit='s').dt.tz_localize('utc').dt.tz_convert('US/Eastern')

# dropping the original 'post_date' column
apple_tweets_df = apple_tweets_df.drop(columns=['post_date'])

# checking the start dates and end dates to verify
display(apple_tweets_df.head())
display(apple_tweets_df.tail())
```

```python
# manipulating the dates for stock data according to NYSE opening and closing time 
# NYSE (Opening time - 9:30 a.m., Closing time - 4:00 p.m. Eastern Time)

# making separate columns for date and time in apple_tweets_df
apple_tweets_df['dates'],apple_tweets_df['time'] = apple_tweets_df['date'].dt.date, apple_tweets_df['date'].dt.hour

apple_tweets_df['updated_dates'] = np.where(apple_tweets_df['time']>=16,apple_tweets_df['date']+datetime.timedelta(days=1),apple_tweets_df['date'])
display(apple_tweets_df)
```

```python
# dropping 'date', 'dates', 'time' and 'writer' column
apple_tweets_dff = apple_tweets_df.drop(columns=['date','dates','time','writer'])
```

```python
# converting the 'updated_dates' column into datetime
apple_tweets_dff['updated_dates'] = pd.to_datetime(apple_tweets_dff['updated_dates'].dt.date)
display(apple_tweets_dff)
```

```python
# viewing the general info of the prepared dataset
apple_tweets_dff.info()
```

## **creating the desired dataframes**

### (1) Same Day aggregated tweets
### (2) Previous Day aggregated tweets
### (3) Last 3 Days aggregated tweets
### (4) Last 7 Days aggregated tweets

### **Same Day aggregated tweets**

```python
# defining a lambda function to combine tweets
combine_tweets = lambda text: " ".join(text)
```

```python
# combining tweets for same day
comb_tweets_apple1 = apple_tweets_dff.groupby(by='updated_dates').agg({'body': combine_tweets,'comment_num':'sum','retweet_num':'sum','like_num':'sum'}).reset_index()
```

```python
display(comb_tweets_apple1)
```

```python
# exporting the dataframe
comb_tweets_apple1.to_csv('/content/gdrive/MyDrive/DATA_602_Project/datasets/Apple/comb_tweets_apple1.csv')
```

## **Last 3 Days aggregated tweets**

```python
# defining column names
col_names = list(comb_tweets_apple1.columns)
print('column names:\n',col_names)

# initializing an empty dataframe with those column names
comb_tweets_apple3d = pd.DataFrame(columns=col_names)
```

```python
# looping over index and combining last 3 rows for each date
for i in comb_tweets_apple1.index:
    if i>1:
        v1 = comb_tweets_apple1['updated_dates'][i]
        v2 = comb_tweets_apple1['body'][i] + ' ' + comb_tweets_apple1['body'][i-1] + ' ' + comb_tweets_apple1['body'][i-2]
        v3 = comb_tweets_apple1['comment_num'][i] + comb_tweets_apple1['comment_num'][i-1] + comb_tweets_apple1['comment_num'][i-2]
        v4 = comb_tweets_apple1['retweet_num'][i] + comb_tweets_apple1['retweet_num'][i-1] + comb_tweets_apple1['retweet_num'][i-2]
        v5 = comb_tweets_apple1['like_num'][i] + comb_tweets_apple1['like_num'][i-1] + comb_tweets_apple1['like_num'][i-2]
        comb_tweets_apple3d = comb_tweets_apple3d.append(pd.DataFrame([[v1,v2,v3,v4,v5]],columns=comb_tweets_apple1.columns),ignore_index=True)
```

```python
display(comb_tweets_apple3d)
```

```python
# exporting the dataframe
comb_tweets_apple3d.to_csv('/content/gdrive/MyDrive/DATA_602_Project/datasets/Apple/comb_tweets_apple3d.csv')
```
